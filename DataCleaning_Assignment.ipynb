```python
#                                       Session DataCleaning 

#****************************************************************************************************
#                                Program 1 : Feeling the missed Flight Number Column and converting to integer
#****************************************************************************************************
# Importing the packages and setting the alias

import numpy as np
import pandas as pd

# Creating a Data Frame of the Data Given 

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',  
                               'Budapest_PaRis', 'Brussels_londOn'], 
              'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 
              'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',  
                               '12. Air France', '"Swiss Air"']}) 
#printing the Data frame
print("The Data given , in which FlightNumber field to be preprocessed \n",'-'*80)
print(df)

# 1. Filling the missing values in FLight number using forward fill , then adding 10 and making it integer 

df['FlightNumber'].fillna(method='ffill',inplace=True)

# Checking if duplicates and correcting the numbers to ascending
df['FlightNumber'] = np.where(df['FlightNumber'].duplicated() ,df['FlightNumber'] + 10,df['FlightNumber'])
df['FlightNumber'] =  pd.to_numeric(df['FlightNumber'] , downcast = 'integer')  


print('-'*80,"\n The Data frame with cleaned Flight Numbers (missed values filled in ascending in 10 and in integer\n",'-'*80)
print(df)

#****************************************************************************************************
```

```python
#****************************************************************************************************
#                                Program 5 : Recent Delays column to be split  
#                                          
#****************************************************************************************************
# Importing the packages and setting the alias

import numpy as np
import pandas as pd

# Creating a Data Frame of the Data Given 

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',  
                               'Budapest_PaRis', 'Brussels_londOn'],
              'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 
              'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
                   'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',  
                               '12. Air France', '"Swiss Air"']}) 
#printing the Data frame
df['FlightNumber'].fillna(method='ffill',inplace=True)
df['FlightNumber'] = np.where(df['FlightNumber'].duplicated() ,df['FlightNumber'] + 10,df['FlightNumber'])
df['FlightNumber'] =  pd.to_numeric(df['FlightNumber'] , downcast = 'integer')  

# creating a Temporary dataFrame and splitting the From to Column
df1 = pd.DataFrame()
df1['Arrival'] = df['From_To'].str.split("_").str.get(0)
df1['Destination'] = df['From_To'].str.split("_").str.get(1)


df1['Arrival'] = df1['Arrival'].str.title()
df1['Destination']  = df1['Destination'].str.title()

df.drop('From_To', inplace=True ,axis=1)

df = pd.concat([df1,df],axis=1)

print(" The Data Frame , in which Recent Delays  to be cleaned\n",'-'*80)
print(df)


# Creating a list of  the Values in the RecentDElays column and splitting into delays
delays = pd.DataFrame(df['RecentDelays'].values.tolist())
delays.columns=['delay_1','delay_2','delay_3']

df = pd.concat([df,delays],axis=1)
df.drop('RecentDelays', inplace=True ,axis=1)

print("\n The Data Frame with Recent Delays  split into multiple columns \n",'-'*80)
print(df)
```
